Now, we create a correlation matrix that measures the linear relationships between the variables. The correlation matrix can be formed by using the corr function from the pandas dataframe library. We will use the heatmap function from the seaborn library to plot the correlation matrix.

```python
# Finding correlation matrix using Heatmap
sns.heatmap(boston.corr(), cmap = 'RdGy')
```

The correlation coefficient ranges from -1 to 1. If the value is close to 1, it means that there is a strong positive correlation between the two variables. When it is close to -1, the variables have a strong negative correlation.

```python
plt.subplots(figsize=(12,8))
sns.heatmap(boston.corr(), annot=True)
```

```python
x=boston[["RM","LSTAT"]]
```

```python
y=boston["MEDV"]
```

```python
# splitting data to training and testing dataset. 
 
#from sklearn.cross_validation import train_test_split
#the submodule cross_validation is renamed and deprecated to model_selection
from sklearn.model_selection import train_test_split
 
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2,
                                                    random_state = 0)
  
print("xtrain shape : ", xtrain.shape)
print("xtest shape  : ", xtest.shape)
print("ytrain shape : ", ytrain.shape)
print("ytest shape  : ", ytest.shape)
```

```python
#Build the model
#Applying Linear Regression Model to the dataset and predicting the prices.  

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#fit the model
model.fit(xtrain, ytrain)
```

```python
# predicting the test set results
y_pred = model.predict(xtest)
y_pred
```

```python
 
# Plotting Scatter graph to show the prediction 
# results - 'ytrue' value vs 'y_pred' value
plt.scatter(ytest, y_pred, c = 'green')
plt.xlabel("Price: in $1000's")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value : Linear Regression")

plt.show()
```

```python
#Display accuracy of a model

from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(ytest, y_pred)
mae = mean_absolute_error(ytest,y_pred)
model_score=model.score(xtest,ytest)
print("Mean Square Error : ", mse)
print("Mean Absolute Error : ", mae)
print("Model Score : ", model_score)
```

```python
# Test the model using user input
features=np.array([[6,19]])
prediction=model.predict(features)
```

```python
#Calculate R2 score
from sklearn.metrics import r2_score
r2= r2_score(ytest,y_pred)
r2
```

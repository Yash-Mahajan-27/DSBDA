```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"C:\Users\91940\Documents\PR_Exam_DSBDA\PR_EXAM_DATASET\PR_EXAM_DATASET\salary.csv")
```

```python
df
```

```python
x=df[['Experience']]
y=df[['Salary']]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
plt.scatter(xtrain,ytrain,color='red')
plt.plot(xtrain,model.predict(xtrain),color='blue')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(xtest,ytest,color='red')
plt.plot(xtrain,model.predict(xtrain),color='blue')
plt.xlabel('Exp')
plt.ylabel('sal')
plt.show
```

```python
#maiking new prediction
new_sal_pred=model.predict([[15]])
print('Predicted Salary of a persons with 2year exp: ',new_sal_pred)
```
